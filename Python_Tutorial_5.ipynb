# Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection. Values must be either integer or floating. For categorical values(features) we will not be able to find out the correlation as categorical features are object types.

```python
df.corr()
```
